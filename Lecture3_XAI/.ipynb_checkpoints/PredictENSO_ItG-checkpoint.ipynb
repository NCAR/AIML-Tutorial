### Input*Gradient Function

```python
def get_gradients(model, inputs, top_pred_idx=None):
    """Computes the gradients of outputs w.r.t input image.

    Args:
        inputs: 2D/3D/4D matrixof samples
        top_pred_idx: (optional) Predicted label for the x_data
                      if classification problem. If regression,
                      do not include.

    Returns:
        Gradients of the predictions w.r.t img_input
    """
    inputs = keras.ops.cast(inputs, tf.float32)

    with tf.GradientTape() as tape:
        tape.watch(inputs)
        
        # Run the forward pass of the layer and record operations
        # on GradientTape.
        preds = model(inputs, training=False)  
        
        # For classification, grab the top class
        if top_pred_idx is not None:
            preds = preds[:, top_pred_idx]
        
    # Use the gradient tape to automatically retrieve
    # the gradients of the trainable variables with respect to the loss.        
    grads = tape.gradient(preds, inputs)
    return grads
```

## Data preparation
Load in the data from the netcdf. This is ERSSTv5 which has been deseasoned, linearly detrended, and has a 5 month lookback running mean applied. From this, we can calculate the Nino3.4 index.

First, we need to include only samples where nino event is occuring (i.e. nino3.4>0.5 or nino3.4<-0.5). Then, we will divide the sst input and nino3.4 output into training, validation and testing chunks.

```python
sst_all = get_data(download=False)

# Preprocess ANN Input Data:
sst = sst_all.where((sst_all.lat <= 30) & (sst_all.lat >= -30), drop=True)
lat = np.asarray(sst.lat)
lon = np.asarray(sst.lon)
time = sst.time

nino34 = get_nino34(sst)

# Grab only El Nino and La Nina samples from input & output
enso_magnitude = 0.5
sst = np.asarray(sst)
x_ann,y_ann = get_nino34events(sst, nino34, enso_magnitude, vectorize=True)
x_cnn,y_cnn = get_nino34events(sst, nino34, enso_magnitude, vectorize=False)

# Split data:
num_samples = y_ann.shape[0]
train_val_test = [0.7,0.85,1] # set train/val/test fraction

xtrain_ann, xval_ann, xtest_ann, ytrain, _, ytest, _, _, _ = split_data(num_samples,
                                                                        train_val_test,
                                                                        x_ann,
                                                                        y_ann)
xtrain_cnn, xval_cnn, xtest_cnn, _, _, _, _, _, _ = split_data(num_samples,
                                                               train_val_test,
                                                               x_cnn,
                                                               y_cnn)

# standardized by the TRAINING STD
xtrain_ann, _, xtest_ann = standardize_data(xtrain_ann,
                                            xval_ann,
                                            xtest_ann)

xtrain_cnn, _, xtest_cnn = standardize_data(xtrain_cnn,
                                            xval_cnn,
                                            xtest_cnn)

# add channel dimension (required for cnn input)
xtrain_cnn = np.expand_dims(xtrain_cnn, axis=1)
xtest_cnn = np.expand_dims(xtest_cnn, axis=1)

#set all nans to zeros (they are learned to be ignored)
xtrain_ann[np.isnan(xtrain_ann)] = 0
xtest_ann[np.isnan(xtest_ann)] = 0

xtrain_cnn[np.isnan(xtrain_cnn)] = 0
xtest_cnn[np.isnan(xtest_cnn)] = 0

# make into categorical format, 1 for El Nino and 0 for La Nina
ytrain = keras.utils.to_categorical(make_cat(ytrain, enso_magnitude))
ytest = keras.utils.to_categorical(make_cat(ytest, enso_magnitude))
```

## Load Trained Neural Network

```python
ann = tf.keras.models.load_model('../Lecture2_ANNvsCNN/saved_model/ENSOmodel_ANN.keras')
cnn = tf.keras.models.load_model('../Lecture2_ANNvsCNN/saved_model/ENSOmodel_CNN.keras')
```

## Plot SST composites of correct predictions
How does the model do on unseen data?

```python
# make predictions for test data
pred_ann = ann.predict(xtest_ann)
pred_cnn = cnn.predict(xtest_cnn) 

pred_ann = np.argmax(pred_ann,axis=1)
pred_cnn = np.argmax(pred_cnn,axis=1)

true = np.argmax(ytest,axis=1)
ntest = true.shape[0]

anncorr = pred_ann==true
cnncorr = pred_cnn==true

nanncorr = anncorr[anncorr].shape[0]
ncnncorr = cnncorr[cnncorr].shape[0]

print('ANN accuracy on testing is %f%%' %(100*nanncorr/ntest))
print('CNN accuracy on testing is %f%%' %(100*ncnncorr/ntest))
```

## Plot SSTs during El Nino and La Nina Predictions

```python
## ANN
elninopred_ann = (pred_ann==1)
laninapred_ann = (pred_ann==0)

elnino_sst_ann = xtest_ann[elninopred_ann,:]
lanina_sst_ann = xtest_ann[laninapred_ann,:]

elnino_sst_ann = np.reshape(elnino_sst_ann,(elnino_sst_ann.shape[0],xtest_cnn.shape[2],xtest_cnn.shape[3]))
lanina_sst_ann = np.reshape(lanina_sst_ann,(lanina_sst_ann.shape[0],xtest_cnn.shape[2],xtest_cnn.shape[3]))

elninoplot_ann = np.mean(elnino_sst_ann,axis=0)
laninaplot_ann = np.mean(lanina_sst_ann,axis=0)

## CNN
elninopred_cnn = (pred_cnn==1)
laninapred_cnn = (pred_cnn==0)

elnino_sst_cnn = xtest_cnn[elninopred_cnn,:]
lanina_sst_cnn = xtest_cnn[laninapred_cnn,:]

elninoplot_cnn = np.mean(elnino_sst_cnn,axis=0)
laninaplot_cnn = np.mean(lanina_sst_cnn,axis=0)
```

```python
## PLOT
projection = ccrs.PlateCarree(central_longitude=180)
transform = ccrs.PlateCarree()

plt.figure(figsize=(8,8))

ax1=plt.subplot(4,1,1,projection=projection)
ax1.pcolormesh(lon,lat,elninoplot_ann,vmin=-2,vmax=2,cmap='RdBu_r',transform=transform)
ax1.coastlines(color='gray')
plt.title('ANN: Composite of El Nino predictions')

ax1=plt.subplot(4,1,2,projection=projection)
ax1.pcolormesh(lon,lat,elninoplot_cnn[0],vmin=-2,vmax=2,cmap='RdBu_r',transform=transform)
ax1.coastlines(color='gray')
plt.title('CNN: Composite of El Nino predictions')

ax2=plt.subplot(4,1,3,projection=projection)
ax2.pcolormesh(lon,lat,laninaplot_ann,vmin=-2,vmax=2,cmap='RdBu_r',transform=transform)
ax2.coastlines(color='gray')
plt.title('ANN: Composite of La Nina predictions')

ax2=plt.subplot(4,1,4,projection=projection)
ax2.pcolormesh(lon,lat,laninaplot_cnn[0],vmin=-2,vmax=2,cmap='RdBu_r',transform=transform)
ax2.coastlines(color='gray')
plt.title('CNN: Composite of La Nina predictions')

plt.show()
```

## Input * Gradient

```python
# Calculate Input*Gradient for correct predictions in each class
# make predictions for test data
pred_ann = ann.predict(xtest_ann)
pred_cnn = cnn.predict(xtest_cnn) 

pred_ann = np.argmax(pred_ann,axis=1)
pred_cnn = np.argmax(pred_cnn,axis=1)

true = np.argmax(ytest,axis=1)

# find where NN is correct
icorr_ann = pred_ann==true
true_corr_ann = true[icorr_ann]
pred_corr_ann = pred_ann[icorr_ann]

icorr_cnn = pred_cnn==true
true_corr_cnn = true[icorr_cnn]
pred_corr_cnn = pred_ann[icorr_cnn]

# split correct predictions into each class (0=la nina, 1=el nino)
ilanina_corr_ann = (pred_corr_ann==0)
ielnino_corr_ann = (pred_corr_ann==1)

ilanina_corr_cnn = (pred_corr_cnn==0)
ielnino_corr_cnn = (pred_corr_cnn==1)

# grab SST input associated with the correct, class-dependent output
sst_lanina_corr_ann = xtest_ann[icorr_ann][ilanina_corr_ann]
sst_elnino_corr_ann = xtest_ann[icorr_ann][ielnino_corr_ann]

sst_lanina_corr_cnn = xtest_cnn[icorr_cnn][ilanina_corr_cnn]
sst_elnino_corr_cnn = xtest_cnn[icorr_cnn][ielnino_corr_cnn]
```

```python
# ------ ANN -------
# calculate input*gradient (ItG)
## arrays of ItG maps
lanina_ItG_heatmap_ann = np.zeros(shape = (np.shape(sst_lanina_corr_ann)[0],xtest_ann.shape[-1]))
elnino_ItG_heatmap_ann = np.zeros(shape = (np.shape(sst_elnino_corr_ann)[0],xtest_ann.shape[-1]))

## arrays of gradient maps
lanina_Grad_heatmap_ann = np.zeros(shape = (np.shape(sst_lanina_corr_ann)[0],xtest_ann.shape[-1]))
elnino_Grad_heatmap_ann = np.zeros(shape = (np.shape(sst_elnino_corr_ann)[0],xtest_ann.shape[-1]))

for s,sample in enumerate(sst_lanina_corr_ann):
    lanina_Grad_heatmap_ann[s] = get_gradients(ann,(sample.reshape((1,xtest_ann.shape[-1]))),top_pred_idx=0).numpy()
    lanina_ItG_heatmap_ann[s] = np.multiply(lanina_Grad_heatmap_ann[s],sample.reshape((1,xtest_ann.shape[-1])))
    
for s,sample in enumerate(sst_elnino_corr_ann):
    elnino_Grad_heatmap_ann[s] = get_gradients(ann,(sample.reshape((1,xtest_ann.shape[-1]))),top_pred_idx=1).numpy()
    elnino_ItG_heatmap_ann[s] = np.multiply(elnino_Grad_heatmap_ann[s],sample.reshape((1,xtest_ann.shape[-1])))

# calculate mean across samples:
lanina_Input_mean_ann = np.reshape(np.mean(sst_lanina_corr_ann, axis=0),(xtest_cnn.shape[2],xtest_cnn.shape[3]))
elnino_Input_mean_ann = np.reshape(np.mean(sst_elnino_corr_ann, axis=0),(xtest_cnn.shape[2],xtest_cnn.shape[3]))

lanina_Grad_meanheatmap_ann = np.reshape(np.mean(lanina_Grad_heatmap_ann, axis=0),(xtest_cnn.shape[2],xtest_cnn.shape[3]))
elnino_Grad_meanheatmap_ann = np.reshape(np.mean(elnino_Grad_heatmap_ann, axis=0),(xtest_cnn.shape[2],xtest_cnn.shape[3]))

lanina_ItG_meanheatmap_ann = np.reshape(np.mean(lanina_ItG_heatmap_ann, axis=0),(xtest_cnn.shape[2],xtest_cnn.shape[3]))
elnino_ItG_meanheatmap_ann = np.reshape(np.mean(elnino_ItG_heatmap_ann, axis=0),(xtest_cnn.shape[2],xtest_cnn.shape[3]))


# ------ CNN -------
## arrays of ItG maps
lanina_ItG_heatmap_cnn = np.zeros(shape = (np.shape(sst_lanina_corr_cnn)[0],xtest_cnn.shape[-2],xtest_cnn.shape[-1]))
elnino_ItG_heatmap_cnn = np.zeros(shape = (np.shape(sst_elnino_corr_cnn)[0],xtest_cnn.shape[-2],xtest_cnn.shape[-1]))

## arrays of gradient maps
lanina_Grad_heatmap_cnn = np.zeros(shape = (np.shape(sst_lanina_corr_cnn)[0],xtest_cnn.shape[-2],xtest_cnn.shape[-1]))
elnino_Grad_heatmap_cnn = np.zeros(shape = (np.shape(sst_elnino_corr_cnn)[0],xtest_cnn.shape[-2],xtest_cnn.shape[-1]))

for s, sample in enumerate(sst_lanina_corr_cnn):
    lanina_Grad_heatmap_cnn[s] = get_gradients(cnn,(sample.reshape((1,1,xtest_cnn.shape[-2],xtest_cnn.shape[-1]))),top_pred_idx=0).numpy()
    lanina_ItG_heatmap_cnn[s] = np.multiply(lanina_Grad_heatmap_cnn[s],sample.reshape((1,xtest_cnn.shape[-2],xtest_cnn.shape[-1])))
    
for s, sample in enumerate(sst_elnino_corr_cnn):
    elnino_Grad_heatmap_cnn[s] = get_gradients(cnn,(sample.reshape((1,1,xtest_cnn.shape[-2],xtest_cnn.shape[-1]))),top_pred_idx=1).numpy()
    elnino_ItG_heatmap_cnn[s] = np.multiply(elnino_Grad_heatmap_cnn[s],sample.reshape((1,xtest_cnn.shape[-2],xtest_cnn.shape[-1])))

# calculate mean across samples:
lanina_Input_mean_cnn = np.mean(sst_lanina_corr_cnn, axis=0)
elnino_Input_mean_cnn = np.mean(sst_elnino_corr_cnn, axis=0)

lanina_Grad_meanheatmap_cnn = np.mean(lanina_Grad_heatmap_cnn, axis=0)
elnino_Grad_meanheatmap_cnn = np.mean(elnino_Grad_heatmap_cnn, axis=0)

lanina_ItG_meanheatmap_cnn = np.mean(lanina_ItG_heatmap_cnn, axis=0)
elnino_ItG_meanheatmap_cnn = np.mean(elnino_ItG_heatmap_cnn, axis=0)
```

```python
# smooth gradient to make plot prettier
# ------ ANN -------
lanina_Grad_meanheatmap_ann = scipy.ndimage.gaussian_filter(lanina_Grad_meanheatmap_ann, sigma=0.7)
elnino_Grad_meanheatmap_ann = scipy.ndimage.gaussian_filter(elnino_Grad_meanheatmap_ann, sigma=0.7)

lanina_ItG_meanheatmap_ann = scipy.ndimage.gaussian_filter(lanina_ItG_meanheatmap_ann, sigma=0.7)
elnino_ItG_meanheatmap_ann = scipy.ndimage.gaussian_filter(elnino_ItG_meanheatmap_ann, sigma=0.7)

# ------ CNN -------
lanina_Grad_meanheatmap_cnn = scipy.ndimage.gaussian_filter(lanina_Grad_meanheatmap_cnn, sigma=0.7)
elnino_Grad_meanheatmap_cnn = scipy.ndimage.gaussian_filter(elnino_Grad_meanheatmap_cnn, sigma=0.7)

lanina_ItG_meanheatmap_cnn = scipy.ndimage.gaussian_filter(lanina_ItG_meanheatmap_cnn, sigma=0.7)
elnino_ItG_meanheatmap_cnn = scipy.ndimage.gaussian_filter(elnino_ItG_meanheatmap_cnn, sigma=0.7)
```
